```python
col1 = widgets.ToggleButtons(
    options=df.columns.tolist(),
    value = 'Educación',
    disabled=False,
    button_style='', layout = Layout(width = '150px', height = '500px'))
col2 = widgets.ToggleButtons(
    options=df.columns.tolist(),
    value = 'Edad',
    disabled=False,
    button_style='', layout = Layout(width = '150px', height = '500px'))
```

```python
graf = widgets.ToggleButtons(
    options = ['Pie', 'Bar', 'Scatter'],
    value = 'Pie',
    disabled=False,
    button_style='', layout = Layout(width = '150px', height = '340px'))
```

```python
def f(col1, col2, graf):
    grupo = col1
    variable = col2
    #print(grupo, variable)
    if (grupo == '') or (variable == ''):
        pass
    else:
        if graf == 'Pie':
            if grupo == variable:
                display()
            else:
                res = pd.pivot_table(df, values = variable, index = grupo, aggfunc='mean').reset_index()
                res = res.sort_values(by = grupo, ascending = True)

                labels = res[grupo]
                valores = res[variable]

                fig = go.Figure()
                # Create subplots: use 'domain' type for Pie subplot
                fig.add_trace(go.Pie(labels=labels, values=valores, name=grupo))

                fig.update_traces(hole=.4, hoverinfo="label+percent+name")
                fig.update_layout(autosize=True,width=400, height=400, title_text='', barmode='group', xaxis_tickangle=45,
                                 margin=dict(l=50, r=0, b=0, t=0, pad=0), template="simple_white",
                                 legend=dict(yanchor="auto", x= 1.55, y=0.8, xanchor="auto", entrywidth=30))
                #FigurA = go.FigureWidget(fig)

                display(fig)
                
        if graf == 'Bar':
            if grupo == variable:
                display()
            else:
                
                res = pd.pivot_table(df, values = variable, index = grupo, aggfunc='mean').reset_index()
                res = res.sort_values(by = grupo, ascending = True)
                res1 = pd.pivot_table(df, values = variable, index = grupo, aggfunc='std').reset_index()
                res1 = res1.sort_values(by = grupo, ascending = True)

                labels = res[grupo].tolist()
                valores = res[variable].tolist()
                valoresDES = res1[variable].tolist()

                fig1 = go.Figure()
                fig1.add_trace(go.Bar(
                    x = labels,
                    y = valores,
                    name='Primary Product',
                    marker_color='salmon', error_y=dict(type='data', color='black', thickness=1.5, width=3,
                                                           array=valoresDES)))

                fig1.update_layout(autosize=True,width=500, height=300, title_text='', barmode='group', xaxis_tickangle=45,
                                 margin=dict(l=50, r=2, b=0, t=15, pad=0), template="simple_white")

                #FigurB = go.FigureWidget(fig1)


                display(fig1)
            
        if graf == 'Scatter':
            if grupo == variable:
                ejex = df[grupo].tolist()
                ejey = df[grupo].tolist()
            else:
                x = df[[grupo, variable]]
                ejex = x[grupo].tolist()
                ejey = x[variable].tolist()
            
            fig2 = go.Figure(data=go.Scatter(x=ejex, y=ejey, mode='markers'))
            
            fig2.update_layout(autosize=True,width=500, height=300, title_text='', barmode='group', xaxis_tickangle=45,
                             margin=dict(l=50, r=2, b=0, t=15, pad=0), template="simple_white",
                              xaxis=dict(title=dict(text=grupo)),
                              yaxis=dict(title=dict(text=variable)))

            #FigurC = go.FigureWidget(fig2)
            display(fig2)
            
    
col_out = widgets.interactive_output(f, {'col1': col1, 'col2': col2, 'graf':graf})
```
