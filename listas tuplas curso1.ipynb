# Las tuplas son una estructura de datos muy útil cuando quieres asegurarte de que los elementos no sean modificados. Son especialmente útiles en aplicaciones como representar puntos en el espacio, almacenar valores constantes, o como claves en un diccionario (dado que las tuplas son inmutables y por lo tanto, pueden usarse como claves, mientras que las listas no) 😊
